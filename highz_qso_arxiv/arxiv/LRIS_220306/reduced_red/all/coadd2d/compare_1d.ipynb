```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits
from astropy.table import Table
```

```python
hdul_coadd2d = fits.open("Science_coadd/spec1d_r220306-r220306-J0901+2906.fits")
hdul_coadd1d = fits.open("../J0901+2906/J0901+2906_coadd_noflux.fits")

output_coadd2d = Table(hdul_coadd2d[2].data)
wave_coadd2d = np.array(output_coadd2d["OPT_WAVE"])
flux_coadd2d = np.array(output_coadd2d["OPT_COUNTS"])

output_coadd1d = Table(hdul_coadd1d[1].data)
wave_coadd1d = np.array(output_coadd1d["wave"])
flux_coadd1d = np.array(output_coadd1d["flux"])
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
mask = ((wave_coadd2d > 7000) & (wave_coadd2d < 11000))
ax.plot(wave_coadd2d[mask], flux_coadd2d[mask], color="black", lw=1, label="1d spectra from coadd2d")
ax.plot(wave_coadd1d, flux_coadd1d, lw=1, alpha=0.6, label="1d spectra from coadd1d")
ax.set_xlim(7400, 10500)
ax.set_ylim(-50, 400)
ax.legend()
```
